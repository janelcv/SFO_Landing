```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
from bs4 import BeautifulSoup
import requests
import time
```

```python
#importing files 
df = "Resources/SF_delay_df.csv"
df = pd.read_csv(df)
df.head()
```

```python
#Creating an array of unique values
unique_tailnum = df["TailNum"].unique()
```

```python
uniquelist = unique_tailnum + " is Not Assigned/Reserved"
```

```python
uniquelist
```

```python
len(uniquelist)
```

```python
#Creating a list
tail_number = unique_tailnum.tolist()
```

```python
tail_number[]
```

```python
#Creating list of urls for each unique Tail Number  
new_urls = ["https://registry.faa.gov/aircraftinquiry/NNum_Results.aspx?nNumberTxt=" + x for x in tail_number]
```

```python
new_urls
```

```python
name_list = []
manufacturer_list = []
model_list = []
year_list = []
tail_list = []
```

```python
#Creating a function which will scrape information from a website   
def scrape():
    counter = 0
    for one_url in new_urls:
        try:
            print(one_url)
            response = requests.get(one_url)
            print(response.status_code) # get status code
            soup = BeautifulSoup(response.text, 'html.parser')
            tail_flag = soup.find("span", class_='Inquiry_InputLabel')
            if(tail_flag.text not in uniquelist):
                tail = tail_flag.text
                tail_list.append(tail)
                print(tail)
                model_flag = soup.find("span", class_='Results_DataText', attrs = {'id':'content_Label7'})
                if(model_flag == None):
                    model = soup.find("span", class_='Results_DataText', attrs = {'id':'content_drptrDeRegAircraft_lbDeRegModel_0'}).text
                else:
                    model = model_flag.text
                model_list.append(model)
                print(model)
                manufacturer_flag = soup.find("span", class_='Results_DataText', attrs = {'id':'content_lbMfrName'})
                if(manufacturer_flag == None):
                    manufacturer = soup.find("span", class_='Results_DataText', attrs = {'id':'content_drptrDeRegAircraft_lbDeRegMfrName_0'}).text
                else:
                    manufacturer = manufacturer_flag.text
                manufacturer_list.append(manufacturer)
                print(manufacturer)
                year_flag = soup.find("span", class_='Results_DataText', attrs = {'id':'content_drptrDeRegAircraft_lbDeRegYearMfr_0'})
                if(year_flag == None):
                    year = soup.find("span", class_='Results_DataText', attrs = {'id':'content_Label17'}).text
                else:
                    year = year_flag.text
                year_list.append(year)
                print(year)
                name_flag = soup.find("span", class_='Results_DataText', attrs = {'id':'content_lbOwnerName'})
                if(name_flag == None):
                    name = soup.find("span", class_='Results_DataText', attrs = {'id':'content_drptrDeRegAircraft_lbDROwnerName_0'}).text
                else:
                    name = name_flag .text
                name_list.append(name)
                print(name)
            counter = counter + 1
            print(str(counter)+" completed")
            time.sleep(0.5)
            print("------------------------------------")
        except:
            print(f"Error. Skipping...")     
```

```python
scrape()
```
